## üìå ÊÄªÁªì‰∏ÄÂè•ËØùÔºö

> `tanh` ÊòØ‰∏Ä‰∏™Âπ≥ÊªëÁöÑ S ÂûãÊøÄÊ¥ªÂáΩÊï∞ÔºåËæìÂá∫ËåÉÂõ¥Âú® -1 Âà∞ 1ÔºåÊØî sigmoid Êõ¥‚ÄúÂ±Ö‰∏≠‚ÄùÔºåÂ∏∏Áî®‰∫éÈöêËóèÂ±Ç„ÄÇ


```python
import torch
import matplotlib.pyplot as plt


plt.rcParams['font.sans-serif'] = ['SimHei']  # Áî®Êù•Ê≠£Â∏∏ÊòæÁ§∫‰∏≠ÊñáÊ†áÁ≠æ
plt.rcParams['axes.unicode_minus'] = False  # Áî®Êù•Ê≠£Â∏∏ÊòæÁ§∫Ë¥üÂè∑

x = torch.linspace(start=-20, end=20, steps=1000)
y = torch.tanh(input=x)

_, axs = plt.subplots(1, 2)
axs[0].plot(x, y)
axs[0].grid()
axs[0].set_title("tanhÊøÄÊ¥ªÂáΩÊï∞")


x = torch.linspace(start=-20, end=20, steps=1000, requires_grad=True)
torch.tanh(x).sum().backward()
axs[1].plot(x.detach().numpy(), x.grad)
axs[1].grid()
axs[1].set_title("tanhÂØºÊï∞")
plt.show()
```
